# Feature Engineering

Dengan tujuan agar data dapat lebih dipahami dengan baik, dibuatlah feature-feature baru berdasarkan data yang sudah ada. Feature-feature baru tersebut adalah sebagai berikut : 

- `person_in_house` = Mewakili jumlah orang di dalam 1 rumah.
- `bedroom_in_house` = Mewakili jumlah kamar tidur di dalam 1 rumah.
- `room_in_house` = Mewakili jumlah ruangan di dalam 1 rumah.

Feature baru dibuat dengan cara membagi kolom yang sesuai dengan kolom lain yang relevan.

```python
# Ada berapa Orang di dalam 1 Rumah
df['person_in_house'] = df['population'] / df['households']

# Ada berapa Kamar Tidur dalam 1 Rumah
df['bedroom_in_house'] = df['total_bedrooms'] / df['households']

# Ada berapa Ruangan dalam 1 Rumah
df['room_in_house'] = df['total_rooms'] / df['households']
df.head()
```

# Handling Missing Value

Handling missing values pada dataset penting dilakukan karena adanya missing values dapat menyebabkan beberapa masalah dalam analisis dan model, antara lain:

1. Bias dalam analisis: Jika missing values tidak ditangani dengan baik, dapat menghasilkan bias dalam analisis. Analisis yang dilakukan dapat berpotensi menghasilkan kesimpulan yang tidak representatif atau salah.

2. Pengaruh pada statistik: Missing values dapat mempengaruhi perhitungan statistik seperti mean, median, atau deviasi standar. Jika missing values tidak diperlakukan, statistik tersebut dapat menjadi bias atau tidak mencerminkan nilai yang sebenarnya dari data.

3. Gangguan pada model: Jika missing values tidak ditangani, model yang dibangun menggunakan dataset tersebut dapat mengalami gangguan atau menghasilkan hasil yang tidak akurat. Banyak algoritma machine learning tidak dapat mengoperasikan dengan missing values, sehingga perlu dilakukan pemrosesan data untuk mengisi atau menghapus missing values sebelum membangun model.

4. Informasi yang hilang: Missing values dapat mengakibatkan kehilangan informasi yang berharga. Jika data yang hilang tersebut memiliki hubungan atau pengaruh penting terhadap target variabel atau variabel lain, maka kehilangan informasi tersebut dapat mengurangi akurasi dan ketepatan model yang dibangun.

5. Pengaruh pada interpretasi: Missing values dapat mempengaruhi interpretasi hasil analisis atau model. Jika variabel dengan missing values memiliki hubungan dengan variabel lain atau variabel target, pengabaian missing values dapat menyebabkan kesalahan interpretasi atau kesalahan dalam membuat keputusan berdasarkan hasil analisis.

Dengan melakukan handling missing values, kita dapat meminimalkan dampak negatif yang disebabkan oleh adanya missing values dalam dataset. Hal ini termasuk pengisian nilai yang hilang, penghapusan baris dengan missing values, atau menggunakan teknik lain seperti imputasi atau metode statistik yang sesuai. Dengan cara ini, data dapat digunakan secara lebih efektif dan menghasilkan hasil analisis atau model yang lebih akurat dan representatif.

# Cek Outlier

Keberadaan outlier dalam dataset dapat memiliki beberapa dampak dan implikasi, antara lain:

1. Bias statistik: Outlier dapat menyebabkan bias statistik yang signifikan. Misalnya, jika kita menghitung rata-rata harga rumah, keberadaan outlier dengan nilai yang sangat tinggi atau rendah dapat secara signifikan dapat mempengaruhi hasil rata-rata yang sebenarnya tidak mencerminkan keadaan umum dari mayoritas data.

2. Performa model: Outlier dapat mempengaruhi performa model statistik atau machine learning yang dibangun menggunakan dataset tersebut. Model cenderung sensitif terhadap nilai ekstrem dan mungkin mencoba untuk menyesuaikan diri dengan outlier, menghasilkan model yang tidak mewakili tren umum data.

3. Bias analisis: Outlier dapat mempengaruhi interpretasi dan kesimpulan dari analisis yang dilakukan. Outlier yang tidak dikendalikan atau tidak diperlakukan dengan benar dapat menyebabkan interpretasi yang salah tentang tren, pola, atau hubungan dalam data.

4. Pengaruh pada prediksi harga: Outlier dapat mempengaruhi prediksi harga rumah yang dibangun menggunakan model machine learning. Jika outlier diabaikan atau tidak dikelola dengan baik, hasil prediksi harga dapat menjadi tidak akurat atau tidak dapat diandalkan.

5. Ketidakterwakilan data: Outlier dapat menyebabkan data yang tidak mewakili populasi secara umum. Jika outlier tersebut merupakan kasus yang jarang terjadi atau kejadian yang ekstrem, maka data tersebut mungkin tidak mewakili karakteristik umum dari pasar properti di California.

# Handling Outlier

Langkah yang akan dilakukan terdiri dari beberapa proses, yakni sebagai berikut : 

- Menghitung IQR (Interquartile Range) pada setiap kolom.
- Tetapkan Batas Atas dan Batas Bawah pada setiap kolom.
- Identifikasi Outlier yang berada diluar Batas Atas maupun Batas Bawah pada Data.
- Eliminasi data yang telah teridentifikasi sebagai Outlier.

```python
df_model.head()
```

```python
def penanganan_outliers(df_model):
    column_list = [  
                   'housing_median_age',
                   'total_rooms',
                   'total_bedrooms',  
                   'population', 
                   'households',
                   'median_income',
                   'median_house_value',
                   'person_in_house', 
                   'bedroom_in_house', 
                   'room_in_house'
                   ]
    bound_dict = {}
    
    for col in column_list:
        if pd.api.types.is_numeric_dtype(df_model[col]):
            
            # Cari IQR and Batas Atas & Bawah
            q1 = df_model[col].quantile(0.25)
            q3 = df_model[col].quantile(0.75)
            iqr = q3 - q1
            lower_bound = q1 - 1.5*iqr
            upper_bound = q3 + 1.5*iqr

            # Simpan batas untuk setiap kolom dalam dictionary
            bound_dict[col] = (lower_bound, upper_bound)

            # Hitung outlier
            under_q1 = df_model[df_model[col] < lower_bound][col].count()
            over_q3 = df_model[df_model[col] > upper_bound][col].count()
            print(f"\n> Number of {col} < Lower Limit ({lower_bound}) = {under_q1} | Number of {col} > Upper Limit ({upper_bound}) = {over_q3}")

        else:
            print(f'\nColumn: {col} is not numeric, skipping.')
            
    # Remove outliers
    for col in column_list:
        if col in bound_dict:
            lower_bound, upper_bound = bound_dict[col]
            df_model = df_model[(df_model[col] >= lower_bound) & (df_model[col] <= upper_bound)]
    
    return df_model
```

# Hyperparameter Tuning pada Catboost

Hyperparameter tuning adalah proses mencari kombinasi terbaik dari hyperparameter dalam model CatBoost untuk mencapai performa yang optimal. Berikut adalah penjelasan singkat tentang hyperparameter yang digunakan:

1. Learning Rate: Menentukan seberapa besar kontribusi setiap pohon dalam model. Learning rate yang lebih kecil memerlukan lebih banyak iterasi untuk mencapai konvergensi, tetapi bisa menghasilkan model yang lebih akurat.

2. Depth: Menentukan kedalaman maksimum dari setiap pohon dalam ensemble. Pohon dengan kedalaman lebih besar dapat menangkap hubungan yang lebih kompleks dalam data, tetapi juga dapat menyebabkan overfitting jika tidak diatur dengan baik.

3. L2 Leaf Regularization: Menerapkan regularisasi L2 pada setiap leaf dalam pohon. Regularisasi ini membantu menghindari overfitting dan meningkatkan generalisasi model.

4. Iterations: Menentukan jumlah maksimum iterasi atau pohon yang akan dibangun. Jumlah iterasi yang lebih tinggi dapat meningkatkan performa model, tetapi juga meningkatkan waktu pelatihan.

5. Subsample: Menentukan persentase sampel yang digunakan untuk membangun setiap pohon. Menggunakan subsample yang lebih kecil dapat membantu mengurangi overfitting dan mempercepat waktu pelatihan.

6. Colsample Bylevel: Menentukan persentase fitur yang digunakan untuk membangun setiap level dalam pohon. Menggunakan colsample bylevel yang lebih kecil dapat mengurangi kompleksitas model dan menghindari overfitting.

Selama proses hyperparameter tuning, berbagai kombinasi nilai hyperparameter ini dieksplorasi untuk menemukan kombinasi yang memberikan performa terbaik pada dataset dan tujuan analisis yang diinginkan. Ini melibatkan penggunaan metode seperti grid search, random search, atau teknik optimasi lainnya untuk menemukan kombinasi hyperparameter optimal.

```python
learning_rate = [0.01, 0.05, 0.1, 0.15, 0.2]
depth = [6, 7, 8, 9, 10]
l2_leaf_reg = [1, 3, 5]
iterations = [10, 100, 500]
subsample = [0.5, 0.6, 0.7, 0.8]
colsample_bylevel = [0.5, 0.6, 0.7, 0.8]

hyperparam_space_catboost = {
    'model__subsample': subsample, 
    'model__learning_rate': learning_rate, 
    'model__l2_leaf_reg': l2_leaf_reg, 
    'model__iterations': iterations, 
    'model__depth': depth, 
    'model__colsample_bylevel': colsample_bylevel
}
```

```python
# Benchmark Model dengan Hyperparameter Tuning
cat = CatBoostRegressor(random_state=42)

estimator_cat = Pipeline([
        ('preprocessing', transformer),
        ('scaler', scaler),
        ('model', cat)
        ])

crossval = KFold(n_splits=5, shuffle=True, random_state=42)

# Hyperparameter tuning
random_cat = RandomizedSearchCV(
    estimator_cat, 
    param_distributions = hyperparam_space_catboost,
    n_iter=100,
    cv = crossval, 
    scoring = ['neg_mean_absolute_error', 'neg_root_mean_squared_error', 'neg_mean_absolute_percentage_error'], 
    n_jobs = -1,
    refit = 'neg_mean_absolute_percentage_error', # Pilih salah satu metric untuk optimalisasi
    random_state = 42  
)
```

Melakukan fitting pada Data Train

```python
random_cat.fit(X_train, y_train)
```

```python
# Hasil Tuning
pd.DataFrame(random_cat.cv_results_).sort_values(by=['rank_test_neg_mean_absolute_error', 'rank_test_neg_root_mean_squared_error', 'rank_test_neg_mean_absolute_percentage_error']).head()
```

```python
# Cek Parameter dan Score terbaik
print('CatBoost Regressor')
print('Best_params:', random_cat.best_params_)
print('Best_score:', random_cat.best_score_)
```

```python
model = {'CatBoost Regressor': CatBoostRegressor(random_state=42)}

# Model dengan estimator terbaik
cat_tuning = random_cat.best_estimator_

# Fitting
cat_tuning.fit(X_train, y_train)

# Predict pada Test Set
y_pred_cat_tuning = cat_tuning.predict(X_test)

mae_cat_tuning = mean_absolute_error(y_test, y_pred_cat_tuning)
rmse_cat_tuning = np.sqrt(mean_squared_error(y_test, y_pred_cat_tuning))
mape_cat_tuning = mean_absolute_percentage_error(y_test, y_pred_cat_tuning)

score_after_tuning_cat = pd.DataFrame({'MAE': mae_cat_tuning, 'RMSE': rmse_cat_tuning, 'MAPE': mape_cat_tuning}, index=model.keys())
score_after_tuning_cat
```

# Hyperparameter Tuning pada Random Forest Regressor

Hyperparameter tuning pada Random Forest Regressor melibatkan penyesuaian nilai dari beberapa hyperparameter berikut untuk mencapai performa yang optimal:

1. Max Depth: Menentukan kedalaman maksimum dari setiap pohon dalam ensemble. Kedalaman yang lebih besar dapat memungkinkan model menangkap pola yang lebih kompleks dalam data, tetapi juga dapat menyebabkan overfitting. Menentukan nilai yang tepat untuk max depth penting untuk mencapai keseimbangan antara underfitting dan overfitting.

2. Min Samples Split: Menentukan jumlah minimum sampel yang diperlukan untuk membagi sebuah node. Jika jumlah sampel di sebuah node tidak memenuhi kriteria ini, pemisahan node tidak akan dilakukan. Menentukan nilai yang lebih tinggi dapat membantu mencegah overfitting dengan membatasi pemisahan node yang terlalu kecil.

3. Min Samples Leaf: Menentukan jumlah minimum sampel yang diperlukan untuk membentuk sebuah leaf (daun) pada pohon. Jika jumlah sampel di sebuah leaf tidak memenuhi kriteria ini, proses pembentukan leaf tidak akan dilakukan. Menentukan nilai yang lebih tinggi dapat membantu mencegah overfitting dengan membatasi leaf yang terlalu kecil.

4. Max Features: Menentukan jumlah fitur yang dipertimbangkan saat mencari pemisahan terbaik pada setiap node. Nilai yang lebih rendah dapat membantu mengurangi kompleksitas model dan menghindari overfitting.

5. n_estimators: Menentukan jumlah pohon yang akan dibangun dalam ensemble. Semakin banyak pohon yang dibangun, semakin kompleks modelnya dan semakin baik kemampuannya untuk mempelajari pola dalam data. Namun, ada batasan pada jumlah pohon yang sebaiknya tidak terlalu besar agar waktu pelatihan tetap dapat dikelola.

Selama proses hyperparameter tuning, berbagai kombinasi nilai hyperparameter ini dieksplorasi dan dievaluasi menggunakan metode seperti grid search, random search, atau teknik optimasi lainnya untuk menemukan kombinasi hyperparameter yang memberikan performa terbaik pada dataset dan tujuan analisis yang diinginkan.

```python
max_depth = list(np.arange(1, 11))
min_samples_split = list(np.arange(2, 21))
min_samples_leaf = list(np.arange(1, 21))
max_features = list(np.arange(2, 11))
n_estimators = list(np.arange(100, 201))

hyperparam_space_rf = {
    'model__max_depth': max_depth,
    'model__min_samples_split': min_samples_split,
    'model__min_samples_leaf': min_samples_leaf,
    'model__max_features': max_features,
    'model__n_estimators': n_estimators
}
```

```python
rf = RandomForestRegressor(random_state=42)

estimator_rf = Pipeline([
    ('preprocessing', transformer),
    ('scaler', scaler),
    ('model', rf)
])

crossval = KFold(n_splits=5, shuffle=True, random_state=42)

# Hyperparameter Tuning
random_rf = RandomizedSearchCV(
    estimator_rf,
    param_distributions = hyperparam_space_rf,
    n_iter = 100,
    cv = crossval,
    scoring = ['neg_mean_absolute_error', 'neg_root_mean_squared_error', 'neg_mean_absolute_percentage_error'], 
    n_jobs = -1,
    refit = 'neg_mean_absolute_percentage_error', # hanya bisa memilih satu metrik untuk optimisasi
    random_state = 42
)
```

Fitting pada Data Train

```python
random_rf.fit(X_train, y_train)
```

```python
# Cek Parameter dan Score terbaik
print('Random Forest Regressor')
print('Best_params:', random_rf.best_params_)
print('Best_score:', random_rf.best_score_)
```

```python
model = {'Random Forest Regressor': RandomForestRegressor(random_state=42)}

# Model dengan estimator terbaik
rf_tuning = random_rf.best_estimator_

# Fitting
rf_tuning.fit(X_train, y_train)

# Predict pada Test Set
y_pred_rf_tuning = rf_tuning.predict(X_test)

mae_rf_tuning = mean_absolute_error(y_test, y_pred_rf_tuning)
rmse_rf_tuning = np.sqrt(mean_squared_error(y_test, y_pred_rf_tuning))
mape_rf_tuning = mean_absolute_percentage_error(y_test, y_pred_rf_tuning)

score_after_tuning_rf = pd.DataFrame({'MAE': mae_rf_tuning, 'RMSE': rmse_rf_tuning, 'MAPE': mape_rf_tuning}, index=model.keys())
score_after_tuning_rf
```
